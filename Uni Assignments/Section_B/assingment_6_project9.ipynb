```python
import math
from re import sub
from time import sleep

def derivative(f, x, h=0.00001):
    try:
        return (f(x + h) - f(x - h)) / (2 * h)
    except ZeroDivisionError:
        return None

def getUserInputsAndParse():
    example = input("Would you like to see some example functions? [Y/N]")
    while example != "Y" and example != "N":
            example = input("Please use Either Y or N (capitals and only Y or N)")
    else:
         pass
    
    if example == "Y":
        print("""Equation 1:   x^2
Equation 2:   sin(x)
Equation 3:  sin(x) + 10x
Equation 4:  x^2 + 5x + 4
Equation 5:  x^3 + 1 + sin(x^2) + (pi / x^10 ) + x**4 * cos(x)
Equation 6 : 1 / (x + 1)
                        """)
        sleep(0.1)
        func = input("What is the function you'd like to calculate the derivative for? (feel free to copy paste from the examples!)")
    elif example == "N":
        func = input("What is the function you'd like to calculate the derivative for? (E.g x^2 + 5x + 4)")

    x = int(input("What value of x? "))

    print(f"You Have given the equation {func}, And want approximate the derivative at {x}")

    math_functions = {
        "sin": "math.sin",
        "cos": "math.cos",
        "tan": "math.tan",
        "log": "math.log",
        "exp": "math.exp",
        "sqrt": "math.sqrt",
        "pi": "math.pi",
        "e": "math.e",
        "^": "**"
    }

    for original, replacement in math_functions.items():
        func = func.replace(original, replacement)

    func = sub(r'(\d)(x)', r'(\1 * \2)', func)
    return lambda x: eval(func, {"math": math, "x": x}), x

func, x = getUserInputsAndParse()
derivative_result = derivative(func, x)
if derivative_result is None:
    print(f"The derivative at x = {x} is undefined.")
else:
    print(f"""The derivative at x = {x} is approximately:
                                {derivative_result}""")
```
