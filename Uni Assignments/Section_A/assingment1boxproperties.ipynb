First we have done most of the leg work. we need to change our equation however for the surface area. 


okay so we have an equation. surface area = 2 * ((height * length) + (width * length) + (width* height))

```python
def box_area(length, width, height):
    return 2 * ((height * length) + (width * length) + (width* height))

box_area(1, 1, 1) # outputs 6 area of each face is 1 and we have 6 faces so correct
box_area(2, 2, 2) # outputs 24 area of each face is 4 and 4 * 6 = 24 so correct
```

This is all we need for this part, we can add a call of this function into the box_volume function

```python
def box_volume(length, width, height):
    volume = length * width * height
    surface_area = box_area(length, width, height)
    print(f"""Given that your box has a- 
            length: {length}, width: {width} and a height: {height}
            The volume of the box is the length:{length} * width:{width} * height{height} = volume
            so the volume of the box is {volume} units squared.
            The surface area of the box is 2 lots of the height * length + width * length + width * height
            therefore the surface area of the box = 2 * (({height} * {length}) + ({width} * {length}) + ({width}* {height})) = {surface_area} units""")
    return volume, surface_area

def box_area(length, width, height):
    return 2 * ((height * length) + (width * length) + (width* height))

for i in range(0, 15, 3):
    box_volume(i, i+1, i+2)
```
